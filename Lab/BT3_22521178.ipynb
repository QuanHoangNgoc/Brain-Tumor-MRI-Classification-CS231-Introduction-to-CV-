### UT

```python
class ut:
  def __init__(self): pass
  def verbose(flag, verbose, func_dir=""):
      if(flag==False): return
      print("__verbose__", func_dir)
      print(verbose)
  def over(val):
      try: ut.verbose(True, (type(val), val.shape, str(sys.getsizeof(val)) + "Bytes"))
      except Exception as e:
          ut.verbose(True, (type(val), "no-shape", str(sys.getsizeof(val)) + "Bytes"))
```

### Load Dataset

```python
import cv2

RESIZE = (224, 224)
def load_data_convert(folder_path):
  x_list, y_list = [], []
  for file_name in os.listdir(folder_path):
    label = file_name[:3]
    path_filename = os.path.join(folder_path, file_name)
    img = cv2.imread(path_filename)
    img = cv2.resize(img, RESIZE)
    img = img.flatten()
    # dataset
    x_list.append(img)
    if(label == 'cat'): y_list.append(1)
    else: y_list.append(0)

  x = np.array(x_list)
  y = np.array(y_list)
  return x, y
```

### Main

```python
COURSE_DIR = '/gdrive/MyDrive/Colab Notebooks/CS231.O21.KHTN.1'
TASK_DIR = os.path.join(COURSE_DIR, 'Baitap', 'Baitap3')
folder_path = os.path.join(TASK_DIR, 'Dataset', 'dogscats_small', 'train')
folder_test_path = os.path.join(TASK_DIR, 'Dataset', 'dogscats_small', 'test')
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.datasets import load_files

# Load tập dữ liệu dogscats_small
X_train, y_train = load_data_convert(folder_path)
ut.over(X_train)
ut.over(y_train)
X_test, y_test = load_data_convert(folder_test_path)
ut.over(X_test)
ut.over(y_test)

# Tiền xử lý dữ liệu nếu cần (ví dụ: chuẩn hóa dữ liệu)
scaler = StandardScaler()

# Huấn luyện và đánh giá mô hình kNN
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_pipeline = make_pipeline(scaler, knn_model)
knn_pipeline.fit(X_train, y_train)
knn_accuracy = accuracy_score(y_test, knn_pipeline.predict(X_test))

# Huấn luyện và đánh giá mô hình Logistic Regression
lr_model = LogisticRegression()
lr_pipeline = make_pipeline(scaler, lr_model)
lr_pipeline.fit(X_train, y_train)
lr_accuracy = accuracy_score(y_test, lr_pipeline.predict(X_test))

print(f'Độ chính xác của kNN trên tập kiểm tra: {knn_accuracy}')
print(f'Độ chính xác của Logistic Regression trên tập kiểm tra: {lr_accuracy}')
```

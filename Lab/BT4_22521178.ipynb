### UT

```python
class ut:
  def __init__(self): pass
  def verbose(flag, verbose, func_dir=""):
      if(flag==False): return
      print("__verbose__", func_dir)
      print(verbose)
  def over(val):
      try: ut.verbose(True, (type(val), val.shape, str(sys.getsizeof(val)) + "Bytes"))
      except Exception as e:
          ut.verbose(True, (type(val), "no-shape", str(sys.getsizeof(val)) + "Bytes"))
```

### Read Dataset

```python
import cv2
from skimage.feature import hog
from skimage import data, color, exposure

RESIZE = (224, 224)

def load_data_convert(folder_path, label):
  x_list, y_list = [], []
  for file_name in os.listdir(folder_path):
    path_filename = os.path.join(folder_path, file_name)
    img = cv2.imread(path_filename)
    img = cv2.resize(img, RESIZE)

    # convert
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Compute HOG features
    fd, hog_image = hog(img, orientations=8, pixels_per_cell=(8, 8),
                    cells_per_block=(1, 1), visualize=True, multichannel=False)
    # training
    ut.over(fd)
    x_list.append(fd)
    if(label == 1): y_list.append(1)
    else: y_list.append(0)

  x = np.array(x_list)
  y = np.array(y_list)
  return x, y
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.datasets import load_files

# Load tập dữ liệu
X_train, y_train = load_data_convert(folder_path_1, label=1)
X_test, y_test = load_data_convert(folder_path_0, label=0)

X, y = np.concatenate((X_train, X_test), axis=0), np.concatenate((y_train, y_test), axis=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
ut.over(X_train)
ut.over(y_train)
ut.over(X_train)
ut.over(y_train)

# Tiền xử lý dữ liệu nếu cần (ví dụ: chuẩn hóa dữ liệu)
scaler = StandardScaler()

# Huấn luyện và đánh giá mô hình kNN
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_pipeline = make_pipeline(scaler, knn_model)
knn_pipeline.fit(X_train, y_train)
knn_accuracy = accuracy_score(y_test, knn_pipeline.predict(X_test))
print(f'Độ chính xác của kNN trên tập kiểm tra: {knn_accuracy}')

# Huấn luyện và đánh giá mô hình Logistic Regression
lr_model = LogisticRegression()
lr_pipeline = make_pipeline(scaler, lr_model)
lr_pipeline.fit(X_train, y_train)
lr_accuracy = accuracy_score(y_test, lr_pipeline.predict(X_test))
print(f'Độ chính xác của Logistic Regression trên tập kiểm tra: {lr_accuracy}')
```
